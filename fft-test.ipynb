# FFT- POC

I am given a signal and I am required to disintegrate that signals into pure signals. DFT is the simple solution to it. Let us first create the Sample Signal on which we want to apply the DFT. The Sample Signal is generated by adding 2 pure signals of different Frequencies and Amplitude.

If I apply FFT on the Sample Signal, I should be able to know the Frequencies and Amplitudes of the constituent pure signals.

```python
import numpy as np
import matplotlib.pyplot as plotter
import matplotlib.pyplot as plt


# How many time points are needed i,e., Sampling Frequency
samplingFrequency   = 100; 

# At what intervals time points are sampled
samplingInterval       = 1 / samplingFrequency;

# Begin time period of the signals
beginTime           = 0;

# End time period of the signals
endTime             = 2; 

# Frequency of the signals
signal1Frequency     = 3;
signal2Frequency     = 7;
 
# Time points
time        = np.arange(beginTime, endTime, samplingInterval);

# Create two sine waves
amplitude1 = np.sin(2*np.pi*signal1Frequency*time)
amplitude2 = np.sin(2*np.pi*signal2Frequency*time) 


plt.xlabel('Time (in second)')
# Increase the default size of the graph
# Set figure width to 12 and height to 9
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 18
fig_size[1] = 3
plt.rcParams["figure.figsize"] = fig_size
plt.title('Pure Signal with Frequency: ' + str(signal1Frequency) + ' Hz') 
plt.plot(time, amplitude1)
plt.show();

plt.title('Pure Signal with Frequency: ' + str(signal2Frequency) + ' Hz') 
plt.plot(time, amplitude2)
plt.show();


# Add the sine waves
amplitude = amplitude1 + amplitude2
plt.title('Mixed Signal with multiple Frequencies: ') 
plt.plot(time, amplitude)
plt.show();
```

Let's say we recieved above mixed signal as an input and we need to find out the constituent pure signals. By Applying the fft we get following signal-

```python
# Frequency domain representation
fourierTransform = np.fft.fft(amplitude)/len(amplitude)           # Normalize amplitude
fourierTransform = fourierTransform[range(int(len(amplitude)/2))] # Exclude sampling frequency

tpCount     = len(amplitude)
values      = np.arange(int(tpCount/2))
timePeriod  = tpCount/samplingFrequency
frequencies = values/timePeriod

# Frequency domain representation
plt.title('Fourier transform depicting the frequency components')
plt.plot(frequencies, abs(fourierTransform))
plt.xlabel('Frequency')
plt.ylabel('Amplitude')

plt.show()
```
